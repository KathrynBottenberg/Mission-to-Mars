```python
#import dependencies
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# Visit the Mars News Website
url = 'https://redplanetscience.com'
browser.visit(url)

# parse
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# scrape title and summary text from landing page of website
articles = html_soup.find('div', id= 'news', class_='container')
print(articles)
```

```python
# store scraped results in a python data structure with 'title' and 'preview' headers
mars_articles = []
rows=articles.find_all('div', class_="col-md-12")

for row in rows:
    article_title = row.find('div', class_='content_title').get_text()
    article_preview = row.find('div', class_='article_teaser_body').get_text()
    title_prev = {}
    title_prev["title"]=article_title
    title_prev["preview"]=article_preview
    mars_articles.append(title_prev)
print(mars_articles)
    
```

```python
# store scraped data in a JSON file
import json

with open('data.json', 'w') as f:     
    json.dump(mars_articles, f)
```

```python
browser.quit()
```
